### 3. Elabore uma estrutura para representar um produto (código, nome, preço). **Crie uma função** para cadastrar 5 produtos. **Crie outra função** para aplicar 10% de aumento no preço do produto e apresente, por meio **de outra função**, todos os dados do produtos cadastrados, após o aumento. Construa uma função para cada opção do menu a seguir:
Menu do Sistema
1. Cadastrar
2. Reajustar
3. Visualizar
4. Sair

Qual opção deseja?

```
class TipoProduto:
    codigo = 0
    nome = ''
    preco = 0.0

def menu():
    print('\nMenu do sistema')
    print('1 - Cadastro de produtos')
    print('2 - Reajuste dos produtos')
    print('3 - Visualização dos produtos')
    print('4 - Sair do sistema')
    opcao = int(input('Qual opção deseja: '))
    return opcao

def cadastrar(vetprod):
    for i in range(3):
        p = TipoProduto() #inicializando as variáveis da class
        p.codigo = int(input('Cadastre o código: '))
        p.nome = input('Cadastre o nome:')
        p.preco = float(input('Cadastre o preço: '))
        vetprod.append(p)
    print('Cadastro realizado com sucesso!')
    return vetprod

def visualizar(vetprod):
    if len(vetprod) > 0:
        print('Código   Nome            Preço')
        for i in range(len(vetprod)):
            print(vetprod[i].codigo,'\t',vetprod[i].nome,'\t',vetprod[i].preco)
    else:
        print('Não há produtos cadastrados!')

def reajustar(vetprod):
    if len(vetprod) > 0:
        for i in range(len(vetprod)):
            vetprod[i].preco += vetprod[i].preco * 10 / 100
        print('Reajuste realizado com sucesso!')
    else:
        print('Não há produtos cadastrados!')
    return vetprod

def main():
    vet_produto = []
    op = menu()
    while op >= 1 and op <= 3:
        if op == 1:
            vet_produto = cadastrar(vet_produto)
        elif op == 2:
            vet_produto = reajustar(vet_produto)
        elif op == 3:
            visualizar(vet_produto)
        op = menu()

main()
```

### 4. Uma escola precisa montar o cadastro geral de seus alunos. Este cadastro deverá conter as seguintes informações por aluno: nome completo, data de nascimento, telefone, endereço e série atual. Levando em conta que esta escola possui no máximo 500 alunos, como você faria para estruturar estas informações num sistema de gerenciamento para a escola? Implemente utilizando estrutura.
```
class tipo_aluno:
    nome = ''
    data = ''
    telefone = 0
    endereco = ''
    serie = 0

def menu():
    print("\n Menu de opções: \n1. Cadastrar alunos \n2. Consulta por nome \n3. Visualizar todos os dados \n4. Sair")
    option = int(input("Qual opção deseja? "))
    return option

def cadastrar(vetor):
    for i in range(1):
        aluno = tipo_aluno()
        aluno.nome = input("Nome do aluno: ")
        aluno.data = input("Data de nascimento do aluno: ")
        aluno.telefone = int(input("Tefefone: "))
        aluno.endereco = input("Endereço: ")
        aluno.serie = int(input("Série: "))
        vetor.append(aluno)
    print("Cadastro feito com sucesso!")
    return vetor

def consultar(vetor):
    if len(vetor) > 0:
        encontrado = False
        name = input("Informe o nome para consulta: ")
        for i in range(len(vetor)):
            if vetor[i].nome == name:
                encontrado = True
                break
    else:
        print("Não há alunos cadastrados!")
    if encontrado:
        print(f"{'Nome':<15} {'Data':<15} {'Telefone':<15} {'Endereço':<15} {'Série':<15}")
        print(f"{vetor[i].nome:<15} {vetor[i].data:<15} {vetor[i].telefone:<15} {vetor[i].endereco:<15} {vetor[i].serie:<15}")
    else:
        print("Aluno não encontrado!")

def visualizar(vetor):
    if len(vetor) > 0:
        print(f"{'Nome':<15} {'Data':<15} {'Telefone':<15} {'Endereço':<15} {'Série':<15}")
        for i in range(len(vetor)):
            print(f"{vetor[i].nome:<15} {vetor[i].data:<15} {vetor[i].telefone:<15} {vetor[i].endereco:<15} {vetor[i].serie:<15}")
    else:
        print("Não há alunos cadastrados!")

def main():
    vetor = []
    op = menu()
    while op > 0 and op <= 4:
        if op == 1:
            cadastrar(vetor)
        elif op == 2:
            consultar(vetor)
        elif op == 3:
            visualizar(vetor)
        elif op == 4:
            print("Programa finalizado!")
            break
        op = menu()

main()
```

### 5. Faça um programa que realize o cadastro de contas bancarias com as seguintes informações: numero da conta, nome do titular e saldo. O banco permitirá o cadastramento de 15 contas. **Crie uma função para cada opção do menu** a seguir. Utilize a estrutura na passagem de parâmetro:
Menu de opções:
1. Cadastrar contas
2. Visualizar todas as contas
3. Consultar por nome
4. Alterar nome e/ou saldo de um número de conta
5. Excluir a conta com menor saldo

Observação:

* No item de menu 4. Alterar nome e/ou saldo de um número de conta, entenda que apenas pode ser alterado o nome e o saldo depois que você pesquisar pelo número da conta.
* No item 5 do menu, encontre o menor saldo dentre todos cadastrados, **depois exclua esta ÚNICA conta.**. O assunto de excluir algo de um vetor foi dado na disciplina de Algoritmo.



```
# Percorra todo os vetor e quando a condição de enontrar um menor saldo for executada, guarde o índice
# Fora do laço de repetição, exclua pelo índice, por exemplo com o comando pop(índice)
class tipo_cliente:
    numero = 0
    nome = ''
    saldo = 0.0

def menu():
    print("\nMenu de opções: \n1. Cadastrar contas \n2. Visualizar todas as contas \n3. Consulta por nome \n4. Alterar nome e/ou saldo de um número de conta \n5. Excluir a conta com menor saldo \n6. Sair")
    option = int(input("Qual a opção desejada? "))
    return option

def cadastrar(vetor):
    for i in range(1):
        cli = tipo_cliente()
        cli.numero = int(input("Número da conta: "))
        cli.nome = input("Nome do cliente: ")
        cli.saldo = float(input("Saldo da conta: "))
        vetor.append(cli)
    print("Cliente cadastrado com sucesso!")
    return vetor

def visualizar(vetor):
    if len(vetor) > 0:
        print(f"{'Número':<15} {'Nome':<15} {'Saldo':<15}")
        for i in range(len(vetor)):
            print(f"{vetor[i].numero:<15} {vetor[i].nome:<15} {vetor[i].saldo:<15}")
    else:
        print("Não há clientes cadastrados!")

def consultar(vetor):
    if len(vetor) > 0:
        encontrado = False
        name = input("Informe o nome para consulta: ")
        for i in range(len(vetor)):
            if vetor[i].nome == name:
                encontrado = True
                break
    else:
        print("Não há clientes cadastrados!")
    if encontrado:
        print(f"{'Número':<15} {'Nome':<15} {'Saldo':<15}")
        print(f"{vetor[i].numero:<15} {vetor[i].nome:<15} {vetor[i].saldo:<15}")
    else:
        print("Cliente não encontrado!")

def alterar(vetor):
    if len(vetor) > 0:
        number = int(input("Informe o número da conta para consulta: "))
        for i in range(len(vetor)):
            if vetor[i].numero == number:
                novo_nome = input("Informe o novo nome (deixe em branco para manter o atual): ")
                novo_saldo = input("Informe o novo saldo (deixe em branco para manter o atual): ")
                if novo_nome:
                    vetor[i].nome = novo_nome
                if novo_saldo:
                    vetor[i].saldo = float(novo_saldo)
        print("Dados atualizados com sucesso!")
    else:
        print("Não há clientes cadastrados!")
    return vetor

def excluir(vetor):
    if len(vetor) > 0:
        for i in range(len(vetor)):
            if i == 0:
                menorSaldo = vetor[i].saldo
                indice_menor = i
            if vetor[i].saldo < menorSaldo:
                menorSaldo = vetor[i].saldo
                indice_menor = i
        vetor.pop(indice_menor)
        print("Dados excluídos com sucesso!")
    else:
        print("Não há clientes cadastrados!")
    return vetor

def main():
    vetor = []
    op = menu()
    while op > 0 and op <= 6:
        if op == 1:
            vetor = cadastrar(vetor)   #usar desta forma quando houver modificação no vetor
        elif op == 2:
            visualizar(vetor)
        elif op == 3:
            consultar(vetor)
        elif op == 4:
            vetor = alterar(vetor)
        elif op == 5:
            vetor = excluir(vetor)
        elif op == 6:
            print("Programa finalizado!")
            break
        op = menu()

main()
```

```
class tipo_cliente:
    cod_cliente = 0
    nome = ''
    fone = 0

class tipo_doc:
    num_doc = 0
    codigo_cliente = 0
    dia_venc = ''
    dia_pag = ''
    valor = 0.0
    juros = 0.0

def menu():
    print("\n** SISTEMA GERENCIADOR DE CLIENTES E DOCUMENTOS ** \n1. Cadastrar clientes \n2. Relatório de clientes \n3. Cadastrar documentos \n4. Relatório de documentos \n5. Excluir clientes sem documentos \n6. Excluir documentos individuais pelo número \n7. Excluir documentos por cliente \n8. Excluir documentos por período \n9. Alterar as informações dos clientes \n10. Mostrar o total de documentos de determinado cliente \n11. Sair")
    option = int(input("Qual opção deseja? "))
    return option

def pesquisar_codCli(vetclientes, cod_cli):
    for i in range(len(vetclientes)):
        if cod_cli == vetclientes[i].cod_cliente:
            return True
    return False

def cadastrar_cli(vetclientes):
    if len(vetclientes) < 15:
        cod_cli = int(input("Código do cliente: "))
        if pesquisar_codCli(vetclientes, cod_cli):
            print("Código ja existente, insira um novo código!")
        else:
            cli = tipo_cliente()
            cli.cod_cliente = cod_cli
            cli.nome = input("Nome do cliente: ")
            cli.fone = int(input("Telefone do cliente: "))
            vetclientes.append(cli)
            print("Cadastro realizado com sucesso!")
    else:
        print("O limite máximo de clientes cadastrados foi alcançado!")
    return vetclientes

def relatorio_cli(vetclientes):
    if len(vetclientes) > 0:
        print(f"\n*****Clientes***** \n{'Código':<15} {'Nome':<15} {'Telefone':<15}")
        for i in range(len(vetclientes)):
            print(f"{vetclientes[i].cod_cliente:<15} {vetclientes[i].nome:<15} {vetclientes[i].fone:<15}")
    else:
        print("Não há clientes cadastrados!")

def pesquisar_codCliDoc(vetclientes, cod_cliDoc):
    for i in range(len(vetclientes)):
        if cod_cliDoc == vetclientes[i].cod_cliente:
            return True
    return False

def cadastrar_doc(vetclientes, vetdocumentos):
    if len(vetdocumentos) < 30:
        cod_cliDoc = int(input("Código do cliente: "))
        if not pesquisar_codCliDoc(vetclientes, cod_cliDoc):
            print("Cliente inexistente, digite um código existente!")
        else:
            doc = tipo_doc()
            doc.codigo_cliente = cod_cliDoc
            doc.num_doc = int(input("Número do documento: "))
            doc.dia_venc = int(input("Dia do vencimento: "))
            doc.dia_pag = int(input("Dia do pagamento: "))
            doc.valor = float(input("Valor: "))
            if doc.dia_pag > doc.dia_venc:
                doc.juros = float(doc.valor * 0.05)
            else:
                doc.juros = 0
            vetdocumentos.append(doc)
            print("Cadastro realizado com sucesso!")
        return vetdocumentos
    else:
        print("O limite de documentos cadastrados foi alcançado!")
    return vetdocumentos

def relatorio_doc(vetdocumentos):
    if len(vetdocumentos) > 0:
        print(f"\n*****Documentos***** \n{'Número':<15} {'Código':<15} {'Dia do vencimento':<20} {'Dia do pagamento':<20} {'Valor':<15} {'Juros':<15}")
        for i in range(len(vetdocumentos)):
            print(f"{vetdocumentos[i].num_doc:<15} {vetdocumentos[i].codigo_cliente:<15} {vetdocumentos[i].dia_venc:<20} {vetdocumentos[i].dia_pag:<20} R${vetdocumentos[i].valor:<13} R${vetdocumentos[i].juros:<15}")
    else:
        print("Não há documentos cadastrados!")

def excluir_cli_s_doc(vetclientes, vetdocumentos):
    if len(vetclientes) > 0 and len(vetdocumentos) > 0:
        codigo_c = int(input("Código do cliente: "))
        doc_encontrado = False
        cli_encontrado = False
        count = 0
        for c in range(len(vetclientes)):
            if codigo_c == vetclientes[c].cod_cliente:
                cli_encontrado = True
                for d in range(len(vetdocumentos)):
                    if codigo_c == vetdocumentos[d].codigo_cliente:
                        doc_encontrado = True
                        break
                    else:
                        count += 1
                        if count == len(vetdocumentos):
                            index = d
        if not cli_encontrado:
            print("Cliente não encontrado!")
        if doc_encontrado:
            print("Este cliente possui documentos!")
        if cli_encontrado and not doc_encontrado:
            vetclientes.pop(index)
            print("Cliente excluído com sucesso!")
    else:
        print("Não há clientes e documentos cadastrados!")
    return vetclientes

def exluir_doc_num(vetdocumentos):
    if len(vetdocumentos) > 0:
        doc_num = int(input("Informe o número do documento: "))
        doc_encontrado = False
        for i in range(len(vetdocumentos)):
            if doc_num == vetdocumentos[i].num_doc:
                doc_encontrado = True
                index = i
                vetdocumentos.pop(index)
                print("Documento excluído com sucesso!")
                break
            else:
                print("Documento não encontrado!")
    else:
        print("Não há documentos cadastrados!")
    return vetdocumentos

def excluir_doc_cli(vetclientes, vetdocumentos):
    if len(vetclientes) > 0 and len(vetdocumentos) > 0:
        codigo_c = int(input("Informe o código do cliente: "))
        cli_encontrado = False
        doc_encontrado = False
        for c in range(len(vetclientes)):
            if codigo_c == vetclientes[c].cod_cliente:
                cli_encontrado = True
                vet_indices = []
                for d in range(len(vetdocumentos)):
                    if codigo_c == vetdocumentos[d].codigo_cliente:
                        doc_encontrado = True
                        vet_indices.append(d)
        if doc_encontrado:
            index = len(vet_indices) - 1
            for i in range(len(vet_indices)):
                vetdocumentos.pop(vet_indices[index])
                index -= 1
            print(f"{len(vet_indices)} documentos foram excluídos com sucesso!")
        if not cli_encontrado:
            print("Cliente não encontrado!")
        if cli_encontrado and not doc_encontrado:
            print("Este cliente não possui documentos cadastrados!")
    else:
        print("Não há clientes e documentos cadastrados!")
    return vetdocumentos

def alterar_cli(vetclientes):
    if len(vetclientes) > 0:
        number = int(input("Informe o código do cliente: "))
        for i in range(len(vetclientes)):
            if vetclientes[i].cod_cliente == number:
                novo_nome = input("Informe o novo nome (deixe em branco para manter o atual): ")
                novo_fone = input("Informe o novo telefone (deixe em branco para manter o atual): ")
                if novo_nome:
                    vetclientes[i].nome = novo_nome
                if novo_fone:
                    vetclientes[i].fone = int(novo_fone)
                print("Dados atualizados com sucesso!")
            else:
                print("Cliente inexistente, digite um código existente!")
    else:
        print("Não há clientes cadastrados!")
    return vetclientes

def visualizar_total_doc_cli(vetclientes, vetdocumentos):
    if len(vetclientes) > 0 and len(vetdocumentos) > 0:
        codigo_c = int(input("Informe o código do cliente: "))
        cli_encontrado = False
        doc_encontrado = False
        for c in range(len(vetclientes)):
            if codigo_c == vetclientes[c].cod_cliente:
                cli_encontrado = True
                indice = []
                for d in range(len(vetdocumentos)):
                    if codigo_c == vetdocumentos[d].codigo_cliente:
                        doc_encontrado = True
                        indice.append(d)
        if doc_encontrado:
            print(f"{'Número':<15} {'Código':<15} {'Vencimento':<15} {'Pagamento':<15} {'Valor':<15} {'Juros':<15}")
            for i in range(len(indice)):
                print(f"{vetdocumentos[indice[i]].num_doc:<15} {vetdocumentos[indice[i]].codigo_cliente:<15} {vetdocumentos[indice[i]].dia_venc:<15} {vetdocumentos[indice[i]].dia_pag:<15} {vetdocumentos[indice[i]].valor:<15} {vetdocumentos[indice[i]].juros:<15}")
        if cli_encontrado and not doc_encontrado:
            print("Este cliente não possui documentos cadastrados!")
        if not cli_encontrado:
            print("Cliente não encontrado!")
    else:
        print("Não há clientes e/ou documentos cadastrados!")

def excluir_doc_per(vetdocumentos):
    if len(vetdocumentos) > 0:
        inicio = int(input("Dia inicial: "))
        final = int(input("Dia final: "))
        encontrado = False
        vet_indices = []
        for i in range(len(vetdocumentos)):
            if vetdocumentos[i].dia_venc >= inicio and vetdocumentos[i].dia_venc <= final:
                encontrado = True
                vet_indices.append(i)
        if encontrado:
            index = len(vet_indices) - 1
            for i in range(len(vet_indices)):
                vetdocumentos.pop(vet_indices[index])
                index -= 1
            print(f"{len(vet_indices)} documentos foram excluídos com sucesso!")
        else:
            print("Não há documentos com vencimento neste período!")
    else:
        print("Não há documentos cadastrados!")
    return vetdocumentos

def main():
    vetclientes = []
    vetdocumentos = []
    op = menu()
    while op > 0 and op <= 11:
        if op == 1:
            vetclientes = cadastrar_cli(vetclientes)
        elif op == 2:
            relatorio_cli(vetclientes)
        elif op == 3:
            vetdocumentos = cadastrar_doc(vetclientes, vetdocumentos)
        elif op == 4:
            relatorio_doc(vetdocumentos)
        elif op == 5:
            vetclientes = excluir_cli_s_doc(vetclientes, vetdocumentos)
        elif op == 6:
            vetdocumentos = exluir_doc_num(vetdocumentos)
        elif op == 7:
            vetdocumentos = excluir_doc_cli(vetclientes, vetdocumentos)
        elif op == 8:
            vetdocumentos = excluir_doc_per(vetdocumentos)
        elif op == 9:
            vetclientes = alterar_cli(vetclientes)
        elif op == 10:
            visualizar_total_doc_cli(vetclientes, vetdocumentos)
        elif op == 11:
            print("Programa finalizado!")
            break
        op = menu()

main()
```
